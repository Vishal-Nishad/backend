# 1. Exception Handling in Python (Core Concept)

Exception handling is about **gracefully managing runtime errors** so your program doesn’t crash unexpectedly.

### Basic Syntax

```python
try:
    # risky code
except SomeException:
    # handle error
else:
    # runs if NO exception
finally:
    # always runs
```

---

## Execution Flow (VERY important for interviews)

| Block     | When it runs                              |
| --------- | ----------------------------------------- |
| `try`     | Always                                    |
| `except`  | Only if exception occurs                  |
| `else`    | Only if NO exception                      |
| `finally` | ALWAYS (even if return/exception happens) |

---

## Example (Practical)

```python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
    else:
        return f"Result = {result}"
    finally:
        print("Execution completed")
```

### Behavior

```python
divide(10, 2)
# Output:
# Execution completed
# Result = 5.0

divide(10, 0)
# Output:
# Execution completed
# Cannot divide by zero
```

---

## Interview Insight

* `else` is rarely used but shows **clean separation of success logic**
* Avoid catching **generic Exception blindly** unless needed

---

# 2. Multiple Exceptions Handling

```python
try:
    x = int("abc")
except ValueError:
    print("Invalid conversion")
except TypeError:
    print("Type error")
```

Or:

```python
except (ValueError, TypeError):
    print("Handled both")
```

---

# 3. finally Keyword (Critical Concept)

## Key Property:

> `finally` **always executes**, even if:

* exception occurs
* return statement is executed
* program tries to exit

---

## Real-world Example: Resource Handling

```python
def read_file():
    file = open("data.txt", "r")
    try:
        return file.read()
    finally:
        file.close()
```

### Why important?

* Prevents **resource leaks**
* Used in:

  * file handling
  * database connections
  * network sockets

---

## Interview Edge Case (VERY IMPORTANT)

```python
def test():
    try:
        return 1
    finally:
        return 2
```

👉 Output: `2`

### Why?

* `finally` overrides return from `try`

---

## Best Practice

Use **context managers (`with`) instead of finally** for resources:

```python
with open("data.txt") as f:
    data = f.read()
```

---

# 4. Raising Exceptions (raise keyword)

You can **manually trigger exceptions**.

---

## Basic Example

```python
def withdraw(balance, amount):
    if amount > balance:
        raise ValueError("Insufficient balance")
    return balance - amount
```

---

## Why raise?

* Enforce business rules
* Fail fast
* Make bugs obvious

---

## Re-raising Exception

```python
try:
    x = int("abc")
except ValueError:
    print("Logging error")
    raise  # re-throws same exception
```

---

# 5. Custom Exceptions (Very Important for Interviews)

## Why?

Built-in exceptions are generic. Real systems need **domain-specific errors**.

---

## Create Custom Exception

```python
class InsufficientBalanceError(Exception):
    pass
```

---

## Use It

```python
def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalanceError("Not enough funds")
    return balance - amount
```

---

## With Additional Data (Advanced)

```python
class InsufficientBalanceError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Balance {balance}, tried {amount}")
```

---

## Handling Custom Exception

```python
try:
    withdraw(100, 200)
except InsufficientBalanceError as e:
    print(e)
```

# 6. Real-World Example (Production-Level Thinking)

### Scenario: API Input Validation

```python
class InvalidAgeError(Exception):
    pass


def register_user(age):
    if age < 18:
        raise InvalidAgeError("User must be 18+")

    return "User registered"


try:
    register_user(16)
except InvalidAgeError as e:
    print("Error:", e)
```

# 7. Exception Chaining (Advanced / Interview Bonus)

```python
try:
    int("abc")
except ValueError as e:
    raise RuntimeError("Conversion failed") from e
```

👉 Keeps original traceback → very useful in debugging

---

# 8. Best Practices (Interview Gold)

### ✅ Do:

* Catch **specific exceptions**
* Use `finally` for cleanup
* Create custom exceptions for business logic
* Log errors before handling

### ❌ Don’t:

* Use bare `except:` (bad practice)
* Suppress exceptions silently
* Overuse exceptions for normal flow

---

# 9. Common Interview Questions

### Q1: Difference between `finally` and `else`?

* `else` → runs when no exception
* `finally` → runs always

---

### Q2: Does `finally` run after return?

👉 Yes, and can override return

---

### Q3: When to use custom exceptions?

👉 When:

* domain-specific logic
* better debugging
* clean architecture

---

### Q4: What happens if exception not caught?

👉 Program terminates with traceback

# 10. Clean Real-World Pattern (What Interviewers Like)

```python
class PaymentError(Exception):
    pass


def process_payment(amount):
    if amount <= 0:
        raise PaymentError("Invalid amount")

    try:
        print("Processing payment...")
        # simulate failure
        raise ConnectionError("Network issue")
    except ConnectionError as e:
        print("Retrying...")
        raise PaymentError("Payment failed") from e
    finally:
        print("Logging transaction")
```

* `raise` → enforce rules
