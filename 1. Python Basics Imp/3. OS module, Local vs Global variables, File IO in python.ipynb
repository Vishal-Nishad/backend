# 1. `os` Module (Backend Practical Use)

## 🔹 Mental Model

Think of `os` as:

> “Bridge between Python code and the operating system”

Used when your backend needs to:

* interact with environment variables
* handle file paths
* manage processes
* work with filesystem

---

## 🔹 High-Frequency Usage (Must Know)

### 1. Environment Variables (VERY IMPORTANT)

Used for configs, secrets, DB URLs.

```python
import os

DB_URL = os.getenv("DATABASE_URL")
DEBUG = os.getenv("DEBUG", "False") == "True"
```

### ✅ Backend Usage

* Django/FastAPI config
* API keys, secrets
* switching environments (dev/prod)

### ❌ Mistake

```python
os.environ["KEY"]  # crashes if missing
```

✔️ Use:

```python
os.getenv("KEY", default)
```

---

### 2. Path Handling (Cross-platform safe)

```python
from pathlib import Path

BASE_DIR = Path(__file__).resolve().parent
file_path = BASE_DIR / "data" / "file.txt"
```

⚠️ Avoid:

```python
"data/file.txt"  # breaks on Windows/Linux differences
```

---

### 3. File/Directory Operations

```python
os.listdir("logs/")
os.makedirs("logs/", exist_ok=True)
```

---

### 4. Process Handling (Rarely used directly in backend)

```python
os.system("ls")
```

❌ Skip in interviews → not backend-relevant
(Use `subprocess` instead if needed)

---

## 🔹 Real Production Example

### Example: Load config securely

```python
import os

class Settings:
    DB_URL = os.getenv("DATABASE_URL")
    REDIS_URL = os.getenv("REDIS_URL")
    SECRET_KEY = os.getenv("SECRET_KEY")
```

Used in:

* FastAPI settings
* Django settings.py

---

## 🔥 Interview Trap

**Q: Why not hardcode configs?**
Answer:

> Because it breaks 12-factor app principles, prevents environment separation, and exposes secrets.

---

## ❌ Skip (Low Value)

* `os.fork()` (rare in Python backend)
* low-level permissions unless DevOps role

# 2. Local vs Global Variables

## 🔹 Mental Model

| Type   | Scope           | Lifetime        |
| ------ | --------------- | --------------- |
| Local  | inside function | during function |
| Global | module level    | entire runtime  |

---

## 🔹 Basics (Must Be Clear)

```python
x = 10  # global

def func():
    x = 5  # local
    print(x)

func()  # 5
print(x)  # 10
```

---

## 🔹 `global` keyword (IMPORTANT)

```python
count = 0

def increment():
    global count
    count += 1
```

---

## 🔹 Backend Reality (VERY IMPORTANT)

### ❗ Avoid global variables in backend

Why?

* not thread-safe
* breaks in multi-worker setups (Gunicorn, Uvicorn)
* state not shared across processes

---

## 🔹 Production Example (BAD)

```python
cache = {}

def get_user(id):
    if id in cache:
        return cache[id]
```

❌ Problem:

* not shared across workers
* memory leaks
* concurrency issues

---

## ✅ Correct Approach

Use:

* Redis
* DB
* external cache

---

## 🔥 Interview Insight

**Q: Are globals safe in FastAPI/Django?**

Answer:

> No. In production, multiple workers/processes run. Each has its own memory, so globals are inconsistent and unsafe for shared state.

---

## 🔹 When Globals Are OK

* constants

```python
API_VERSION = "v1"
```

* config values (read-only)

---

## ❌ Skip

* deep compiler scope theory (LEGB is enough)

# 3. File I/O (Backend Critical)

## 🔹 Mental Model

> File I/O = persistent data interaction (logs, uploads, temp storage)

---

## 🔹 Core Pattern (Must Know)

```python
with open("file.txt", "r") as f:
    data = f.read()
```

### Why `with`?

* auto close
* prevents memory leaks

---

## 🔹 Modes

| Mode | Meaning   |
| ---- | --------- |
| `r`  | read      |
| `w`  | overwrite |
| `a`  | append    |
| `b`  | binary    |

---

## 🔹 Writing

```python
with open("log.txt", "a") as f:
    f.write("New log\n")
```

---

## 🔹 Backend Use Cases

### 1. Logging (Important)

```python
import logging

logging.basicConfig(filename="app.log", level=logging.INFO)
logging.info("User logged in")
```

---

### 2. File Upload Handling (FastAPI example)

```python
from fastapi import UploadFile

async def upload(file: UploadFile):
    content = await file.read()
    with open(file.filename, "wb") as f:
        f.write(content)
```

---

### 3. Streaming Large Files (IMPORTANT)

❌ Bad:

```python
data = f.read()  # loads entire file
```

✔️ Good:

```python
for chunk in f:
    process(chunk)
```

---

## 🔥 Production Example

### Example: CSV processing (memory-safe)

```python
def process_file(path):
    with open(path, "r") as f:
        for line in f:
            process(line)
```

---

## 🔥 Interview Traps

### 1. File Not Closed

```python
f = open("file.txt")
# forgot close
```

✔️ Use `with`

---

### 2. Encoding Issues

```python
open("file.txt", encoding="utf-8")
```

---

### 3. Concurrent Writes (IMPORTANT)

Multiple workers writing same file → corruption

✔️ Use:

* logging module
* external storage (S3, DB)

---

## 🔹 When NOT to Use File I/O

* storing user data → use DB
* caching → use Redis
* high-concurrency writes → avoid local files

# 🔥 High-Probability Interview Questions

## Q1: Why are global variables dangerous in backend apps?

**Answer:**

* not thread-safe
* not shared across processes
* leads to inconsistent state in multi-worker setups

---

## Q2: How do you manage configuration in production?

**Answer:**

* environment variables via `os.getenv`
* optionally `.env` files
* never hardcode secrets

---

## Q3: How do you handle large file processing?

**Answer:**

* use streaming (`for line in file`)
* avoid loading entire file into memory

---

## Q4: Difference between `os.getenv` and `os.environ`?

**Answer:**

* `getenv` → safe, returns default
* `environ` → raises error if missing
