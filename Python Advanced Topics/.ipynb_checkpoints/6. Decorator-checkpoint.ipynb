Decorators are a **very common Python interview topic**, and they are best understood if you connect three ideas:

1. **Functions are objects in Python**
2. **Functions can receive other functions as arguments**
3. **Functions can return other functions**

A **decorator is simply a function that wraps another function to modify or extend its behavior without changing the original function’s code.**

---

# 1. First Principle: Functions Are Objects

In Python, functions behave like normal objects.

```python
def greet():
    print("Hello")

say_hello = greet
say_hello()
```

Output:

```
Hello
```

Here:

* `say_hello` now **references the same function** as `greet`.

This property enables decorators.

---

# 2. Functions Can Accept Other Functions

Example:

```python
def greet():
    print("Hello Vishal")

def call_function(func):
    func()

call_function(greet)
```

Here:

* `greet` is **passed as an argument**.

This concept is called **higher-order functions**.

---

# 3. Functions Can Return Functions

Example:

```python
def outer():
    def inner():
        print("Inside inner function")
    return inner

func = outer()
func()
```

Output:

```
Inside inner function
```

Here:

* `outer()` returns `inner`.

This ability is **critical for decorators**.

---

# 4. Now Combine These Concepts → Decorator

A decorator:

1. Takes a function
2. Adds extra behavior
3. Returns a new function

Example:

```python
def my_decorator(func):

    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")

    return wrapper
```

Now apply it.

```python
def say_hello():
    print("Hello!")

say_hello = my_decorator(say_hello)

say_hello()
```

Output:

```
Before function execution
Hello!
After function execution
```

What happened?

1. `say_hello` was passed to `my_decorator`
2. `wrapper` was returned
3. `say_hello` now refers to `wrapper`

---

# 5. Decorator Syntax (`@`)

Python provides a cleaner syntax.

Instead of writing:

```python
say_hello = my_decorator(say_hello)
```

You write:

```python
@my_decorator
def say_hello():
    print("Hello!")
```

This is **exactly equivalent**.

Execution becomes:

```
say_hello = my_decorator(say_hello)
```

---

# 6. Real Practical Example (Interview Important)

### Example: Logging function execution

```python
def log_decorator(func):

    def wrapper():
        print(f"Executing function: {func.__name__}")
        func()
        print("Execution finished")

    return wrapper
```

Use it:

```python
@log_decorator
def process_data():
    print("Processing data...")

process_data()
```

Output:

```
Executing function: process_data
Processing data...
Execution finished
```

This is used in **real systems for:**

* logging
* authentication
* timing
* caching
* permissions

---

# 7. Decorator with Arguments (Important for Interviews)

Problem:

The previous decorator only works with functions **without parameters**.

Solution → use `*args, **kwargs`.

```python
def my_decorator(func):

    def wrapper(*args, **kwargs):
        print("Before execution")
        result = func(*args, **kwargs)
        print("After execution")
        return result

    return wrapper
```

Usage:

```python
@my_decorator
def add(a, b):
    return a + b

print(add(3, 4))
```

Output:

```
Before execution
After execution
7
```

---

# 8. Real Production Example → Timing a Function

Very popular interview example.

```python
import time

def timer(func):

    def wrapper(*args, **kwargs):
        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()
        print("Execution time:", end - start)

        return result

    return wrapper
```

Usage:

```python
@timer
def slow_function():
    time.sleep(2)

slow_function()
```

Output:

```
Execution time: 2.00
```

Used in:

* performance measurement
* debugging slow APIs

---

# 9. Very Important Interview Question

### Why do we use decorators?

Decorators allow us to **add functionality without modifying original code**.

Example uses:

| Use Case       | Example                 |
| -------------- | ----------------------- |
| Logging        | log API calls           |
| Authentication | check user permissions  |
| Timing         | measure performance     |
| Caching        | speed up repeated calls |
| Validation     | check inputs            |

Framework examples:

* **Flask**

```python
@app.route("/home")
def home():
    return "Hello"
```

Here `@app.route` is a **decorator**.

* **FastAPI**

```python
@app.get("/users")
```

Also a decorator.

---

# 10. Visual Understanding (Very Important)

Without decorator:

```
main()
  ↓
say_hello()
```

With decorator:

```
main()
  ↓
wrapper()
  ↓
before code
  ↓
say_hello()
  ↓
after code
```

So decorator **wraps the original function**.

---

# 11. One More Advanced Interview Point

When we decorate a function:

```python
@decorator
def func():
```

Actually becomes:

```
func = decorator(func)
```

This line is **extremely important for interviews**.

---

# 12. Short Interview Definition

You can answer like this:

> A decorator in Python is a higher-order function that takes another function as input, adds additional behavior to it, and returns a modified function without changing the original function's code.
