Decorators are a **very common Python interview topic**, and they are best understood if you connect three ideas:

1. **Functions are objects in Python**
2. **Functions can receive other functions as arguments**
3. **Functions can return other functions**

A **decorator is simply a function that wraps another function to modify or extend its behavior without changing the original function’s code.**

---

# 1. First Principle: Functions Are Objects

In Python, functions behave like normal objects.

```python
def greet():
    print("Hello")

say_hello = greet
say_hello()
```

Output:

```
Hello
```

Here:

* `say_hello` now **references the same function** as `greet`.

This property enables decorators.

---

# 2. Functions Can Accept Other Functions

Example:

```python
def greet():
    print("Hello Vishal")

def call_function(func):
    func()

call_function(greet)
```

Here:

* `greet` is **passed as an argument**.

This concept is called **higher-order functions**.

---

# 3. Functions Can Return Functions

Example:

```python
def outer():
    def inner():
        print("Inside inner function")
    return inner

func = outer()
func()
```

Output:

```
Inside inner function
```

Here:

* `outer()` returns `inner`.

This ability is **critical for decorators**.

---

# 4. Now Combine These Concepts → Decorator

A decorator:

1. Takes a function
2. Adds extra behavior
3. Returns a new function

Example:

```python
def my_decorator(func):

    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")

    return wrapper
```

Now apply it.

```python
def say_hello():
    print("Hello!")

say_hello = my_decorator(say_hello)

say_hello()
```

Output:

```
Before function execution
Hello!
After function execution
```

What happened?

1. `say_hello` was passed to `my_decorator`
2. `wrapper` was returned
3. `say_hello` now refers to `wrapper`

---

# 5. Decorator Syntax (`@`)

Python provides a cleaner syntax.

Instead of writing:

```python
say_hello = my_decorator(say_hello)
```

You write:

```python
@my_decorator
def say_hello():
    print("Hello!")
```

This is **exactly equivalent**.

Execution becomes:

```
say_hello = my_decorator(say_hello)
```

---

# 6. Real Practical Example (Interview Important)

### Example: Logging function execution

```python
def log_decorator(func):

    def wrapper():
        print(f"Executing function: {func.__name__}")
        func()
        print("Execution finished")

    return wrapper
```

Use it:

```python
@log_decorator
def process_data():
    print("Processing data...")

process_data()
```

Output:

```
Executing function: process_data
Processing data...
Execution finished
```

This is used in **real systems for:**

* logging
* authentication
* timing
* caching
* permissions

---

# 7. Decorator with Arguments (Important for Interviews)

Problem:

The previous decorator only works with functions **without parameters**.

Solution → use `*args, **kwargs`.

```python
def my_decorator(func):

    def wrapper(*args, **kwargs):
        print("Before execution")
        result = func(*args, **kwargs)
        print("After execution")
        return result

    return wrapper
```

Usage:

```python
@my_decorator
def add(a, b):
    return a + b

print(add(3, 4))
```

Output:

```
Before execution
After execution
7
```

---

# 8. Real Production Example → Timing a Function

Very popular interview example.

```python
import time

def timer(func):

    def wrapper(*args, **kwargs):
        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()
        print("Execution time:", end - start)

        return result

    return wrapper
```

Usage:

```python
@timer
def slow_function():
    time.sleep(2)

slow_function()
```

Output:

```
Execution time: 2.00
```

Used in:

* performance measurement
* debugging slow APIs

# 9. Very Important Interview Question

### Why do we use decorators?

Decorators allow us to **add functionality without modifying original code**.

Example uses:

| Use Case       | Example                 |
| -------------- | ----------------------- |

| Authentication | check user permissions  |
| Timing         | measure performance     |

Framework examples:

* **Flask**

```python

```

Here `@app.route` is a **decorator**.

* **FastAPI**

```python
@app.get("/users")
```

Also a decorator.

Without decorator:

With decorator:

# 1. Why `*args` and `**kwargs` Are Needed in Decorators

First look at a **simple decorator**:

```python
def my_decorator(func):

    def wrapper():
        print("Before execution")
        func()
        print("After execution")

    return wrapper
```

This works **only if the function takes no arguments**.

Example:

```python
@my_decorator
def greet():
    print("Hello")
```

Works fine.

---

## Problem

If the function takes arguments:

```python
@my_decorator
def add(a, b):
    return a + b
```

When we call:

```python
add(2,3)
```

Python actually executes:

```
wrapper(2,3)
```

But our wrapper is:

```python
def wrapper():
```

So Python throws:

```
TypeError: wrapper() takes 0 positional arguments but 2 were given
```

Because the wrapper **cannot accept the arguments**.

---

# 2. Solution → `*args` and `**kwargs`

To support **any function signature**, we write:

```python
def my_decorator(func):

    def wrapper(*args, **kwargs):
        print("Before execution")

        result = func(*args, **kwargs)

        print("After execution")

        return result

    return wrapper
```

Now the wrapper **accepts any arguments**.

---

# 3. What Exactly Are `*args` and `**kwargs`

They are **argument collectors**.

### `*args` → collects positional arguments

Example:

```python
def test(*args):
    print(args)

test(1,2,3)
```

Output

```
(1,2,3)
```

Internally Python stores them as a **tuple**.

---

### `**kwargs` → collects keyword arguments

Example:

```python
def test(**kwargs):
    print(kwargs)

test(a=1,b=2)
```

Output

```
{'a':1,'b':2}
```

Internally stored as a **dictionary**.

---

# 4. How It Works Inside a Decorator

Suppose we have:

```python
@my_decorator
def add(a, b):
    return a + b
```

Calling:

```
add(3,4)
```

is actually:

```
wrapper(3,4)
```

Inside wrapper:

```
args = (3,4)
kwargs = {}
```

Then:

```
func(*args, **kwargs)
```

becomes:

```
func(3,4)
```

which calls the original function.

---

# 5. Example with Both Types

```python
@my_decorator
def student(name, age=20):
    print(name, age)

student("Vishal", age=22)
```

Execution:

```
wrapper("Vishal", age=22)
```

Inside wrapper:

```
args = ("Vishal",)
kwargs = {"age":22}
```

Then:

```
func(*args, **kwargs)
```

becomes

```
student("Vishal", age=22)
```

---

# 6. Are Both `*args` and `**kwargs` Required?

Not always.

### Case 1 — Only positional arguments

```python
def wrapper(*args):
```

works.

Example:

```
add(2,3)
```

---

### Case 2 — Only keyword arguments

```python
def wrapper(**kwargs):
```

works.

Example:

```
add(a=2,b=3)
```

---

### Best Practice

In decorators we almost always write:

```python
def wrapper(*args, **kwargs)
```

Because the decorator **should support any function signature**.

Example functions:

```
func()
func(1,2)
func(a=5)
func(1,2,a=5)
```

Your decorator will work with **all of them**.

This is why **production decorators always use both**.

---

# 7. Important Flow of a Decorator

When you write:

```python
@decorator
def add(a,b):
```

Python converts it to:

```
add = decorator(add)
```

Now:

```
add → wrapper
```

Execution:

```
add(3,4)
↓
wrapper(3,4)
↓
func(*args,**kwargs)
↓
original add(3,4)
```

# 8. Now Let's Understand Class-Based Decorators

Most people get confused here.

But the concept is simple.

A **class decorator works using `__call__()`**.

Objects with `__call__` behave like functions.

---

# 9. First Understand Callable Objects

Example:

```python
class Test:

    def __call__(self):
        print("Object called like function")

t = Test()

t()
```

Output

```
Object called like function
```

Because Python internally does:

```
t()  →  t.__call__()
```

---

# 10. Class Based Decorator Structure

```python
class MyDecorator:

    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("Before execution")

        result = self.func(*args, **kwargs)

        print("After execution")

        return result
```

Usage:

```python
@MyDecorator
def greet():
    print("Hello")

greet()
```

---

# 11. What Python Actually Does

When Python sees:

```python
@MyDecorator
def greet():
```

It converts it to:

```
greet = MyDecorator(greet)
```

So now:

```
greet → instance of MyDecorator
```

---

# 12. Execution Flow

Call:

```
greet()
```

Python does:

```
greet.__call__()
```

Execution steps:

### Step 1

Decorator applied

```
greet = MyDecorator(greet)
```

### Step 2

Object created

```
self.func = greet
```

### Step 3

Function called

```
greet()
```

becomes

```
MyDecorator.__call__()
```

### Step 4

Inside `__call__`

```
print("Before")
self.func()
print("After")
```

---

# 13. Full Example with Arguments

```python
class Logger:

    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):

        print("Function called:", self.func.__name__)

        result = self.func(*args, **kwargs)

        print("Function finished")

        return result
```

Usage:

```python
@Logger
def add(a,b):
    return a+b

print(add(3,4))
```

Output

```
Function called: add
Function finished
7
```

---

# 14. Why Class Decorators Are Used

They are useful when we need:

### 1. State

Example: counting calls.

### 2. Complex logic

### 3. Configurable decorators

Example:

```
@rate_limit(5)
@retry(3)
@cache
```

Many frameworks use class decorators internally.

---

# 15. Example — Count Function Calls

```python
class CountCalls:

    def __init__(self, func):
        self.func = func
        self.count = 0

    def __call__(self, *args, **kwargs):

        self.count += 1
        print("Called", self.count, "times")

        return self.func(*args, **kwargs)
```

Usage:

```python
@CountCalls
def say_hello():
    print("Hello")

say_hello()
say_hello()
say_hello()
```

Output:

```
Called 1 times
Hello
Called 2 times
Hello
Called 3 times
Hello
```

This is **harder to do with function decorators**.

# 17. Interview Tip (Important)

If asked:

**"How does a class decorator work?"**

Answer:

> When a class decorator is applied, Python instantiates the class with the decorated function as an argument to `__init__`. The instance replaces the function. When the function is called, Python invokes the `__call__` method of the instance.
