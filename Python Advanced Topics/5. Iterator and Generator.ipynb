For interviews, **Iterators and Generators** are usually asked to test your understanding of **Python iteration protocol, memory efficiency, and lazy evaluation**. I'll explain them in a **practical + interview-ready way**.

---

# 1. What is Iteration in Python?

Iteration means **accessing elements of a collection one by one**.

Example:

```python
nums = [10, 20, 30]

for n in nums:
    print(n)
```

The `for` loop internally **does not access the list directly**.
Instead, Python converts the list into an **iterator**.

Internal behavior:

```
for item in iterable:
```

Python does roughly this:

```python
iterator = iter(iterable)

while True:
    try:
        item = next(iterator)
        # loop body
    except StopIteration:
        break
```

So to understand generators, you **must understand iterators first**.

---

# 2. What is an Iterator?

An **iterator** is an object that allows you to traverse elements **one at a time**.

An object is an iterator if it implements two methods:

```
__iter__()
__next__()
```

### Responsibilities

| Method       | Purpose                              |
| ------------ | ------------------------------------ |
| `__iter__()` | Returns the iterator object          |
| `__next__()` | Returns next value                   |
|              | Raises `StopIteration` when finished |

---

# 3. Example of an Iterator (Manual Implementation)

Let's create our own iterator.

### Example: Counter

```python
class Counter:
    
    def __init__(self, max):
        self.max = max
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.max:
            self.current += 1
            return self.current
        else:
            raise StopIteration
```

### Usage

```python
c = Counter(3)

for i in c:
    print(i)
```

Output

```
1
2
3
```

---

### Step-by-Step Execution

Python does this internally:

```
c = Counter(3)

iterator = iter(c)  -> calls __iter__()

next(iterator) -> __next__() -> 1
next(iterator) -> __next__() -> 2
next(iterator) -> __next__() -> 3
next(iterator) -> StopIteration
```

---

### Interview Point

**Iterable vs Iterator**

| Term     | Meaning                            |
| -------- | ---------------------------------- |
| Iterable | Object that can return an iterator |
| Iterator | Object that produces next value    |

Examples:

```
Iterable: list, tuple, set, dict, string
Iterator: object returned by iter()
```

Example:

```python
nums = [1,2,3]

it = iter(nums)   # iterator

print(next(it))
print(next(it))
print(next(it))
```

# 4. Problem With Iterators

Writing iterators manually requires:

* writing classes
* managing state
* raising StopIteration

This is **verbose and error-prone**.

Python provides a simpler tool:

# Generators

---

# 5. What is a Generator?

A **generator is a simpler way to create iterators**.

Instead of writing `__next__`, we use the keyword:

```
yield
```

A function containing `yield` becomes a **generator function**.

---

# 6. Example of Generator

### Same Counter Example

```python
def counter(max):

    current = 0
    
    while current < max:
        current += 1
        yield current
```

Usage

```python
for i in counter(3):
    print(i)
```

Output

```
1
2
3
```

---

# 7. How Generator Works Internally

When Python sees `yield`:

```
function → becomes generator object
```

Example

```python
g = counter(3)
```

`g` is **not executed yet**.

Type check:

```python
print(type(g))
```

Output

```
<class 'generator'>
```

Execution happens only when:

```
next(g)
```

---

### Step-by-Step Execution

```
g = counter(3)
```

State:

```
current = 0
```

First call:

```
next(g)
```

Execution:

```
current = 1
yield 1
```

Function **pauses here**.

Second call:

```
next(g)
```

Resume from pause:

```
current = 2
yield 2
```

Third call:

```
current = 3
yield 3
```

Fourth call:

```
StopIteration
```

Key idea:

**Generator remembers its state between calls.**

---

# 8. Generator vs Normal Function

### Normal Function

```python
def test():
    return 1
```

Execution

```
call -> execute -> return -> finish
```

---

### Generator Function

```python
def test():
    yield 1
```

Execution

```
call -> pause -> resume -> pause -> resume
```

Generators support **pause and resume execution**.

# 9. Generator Expressions

Like list comprehension.

### List Comprehension

```python
nums = [x*x for x in range(5)]
```

Output

```
[0,1,4,9,16]
```

Memory:

```
Stores all values
```

---

### Generator Expression

```python
nums = (x*x for x in range(5))
```

Memory:

```
Stores nothing
Calculates on demand
```

Usage

```python
for n in nums:
    print(n)
```

---

# 10. Memory Efficiency (Important Interview Point)

### List

```python
nums = [x*x for x in range(10000000)]
```

Memory:

```
Stores 10 million values
```

---

### Generator

```python
nums = (x*x for x in range(10000000))
```

Memory:

```
Stores only ONE value at a time
```

This is called:

**Lazy Evaluation**

# 11. Real World Use Cases

### 1. Reading Large Files

Without generator

```python
lines = file.readlines()
```

Problem:

```
Loads entire file in memory
```

---

Using generator

```python
def read_file(file):

    for line in file:
        yield line
```

Memory efficient.

---

### 2. Data Pipelines

Example:

```
log file → filter errors → process → send
```

Generators allow **stream processing**.

---

### 3. Infinite Sequences

Example:

```python
def infinite_counter():

    i = 0

    while True:
        yield i
        i += 1
```

Usage

```python
g = infinite_counter()

print(next(g))
print(next(g))
print(next(g))
```

# 12. Key Differences (Interview Ready)

| Feature        | Iterator                                          | Generator              |
| -------------- | ------------------------------------------------- | ---------------------- |
| Definition     | Object implementing `__iter__()` and `__next__()` | Function using `yield` |
| Complexity     | More code                                         | Very simple            |
| State Handling | Manual                                            | Automatic              |
| Memory         | Efficient                                         | Efficient              |
| Creation       | Class                                             | Function               |

# 13. Important Interview Questions

### Q1: Are generators iterators?

Yes.

Generator objects **implement iterator protocol**.

```
__iter__()
__next__()
```

---

### Q2: Can generator be iterated twice?

No.

Example

```python
g = (x for x in range(3))

list(g)
list(g)
```

Output

```
[0,1,2]
[]
```

Generator is **exhausted after one iteration**.

---

### Q3: Difference between `yield` and `return`

| return            | yield                    |
| ----------------- | ------------------------ |
| Ends function     | Pauses function          |
| Returns one value | Produces multiple values |

---

# 14. Interview Summary (Very Important)

You can answer like this:

> An **iterator** is an object that implements the Python iterator protocol using `__iter__()` and `__next__()` to produce values one at a time.
>
> A **generator** is a simpler way to create iterators using the `yield` keyword.
>
> Generators provide **lazy evaluation**, meaning values are generated only when needed, making them **memory efficient** for large data processing.
