# 1. Step 1 — Core idea (lock this first)

## Instance attribute

👉 Belongs to **each object separately**

## Class attribute

👉 Shared across **all objects (by default)**

---

# 2. Step 2 — See it with minimal code

```python
class User:
    role = "customer"   # class attribute

    def __init__(self, name):
        self.name = name  # instance attribute
```

### Objects

```python
u1 = User("A")
u2 = User("B")
```

### Result

* `u1.name` → "A"

* `u2.name` → "B"
  ✔ different (instance-specific)

* `u1.role` → "customer"

* `u2.role` → "customer"
  ✔ same (shared from class)

---

# 3. Step 3 — Memory trick (important)

👉 **Instance attribute = "this object’s data"**
👉 **Class attribute = "default/shared config"**

---

# 4. Step 4 — What actually happens internally (interview gold)

When you do:

```python
u1.role
```

Python checks:

1. Does `u1` have `role`? ❌
2. Then check class → `User.role` ✔

So it finds class attribute.

---

# 5. Step 5 — Overriding behavior (very common question)

```python
u1.role = "admin"
```

Now:

* `u1.role` → "admin"
* `u2.role` → "customer"

### Why?

Because now:

* `u1` has its **own instance attribute**
* It **shadows** the class attribute

# 6. Step 6 — Real backend example (important)

## Class attribute = config

```python
class APIClient:
    timeout = 5  # class attribute

    def __init__(self, base_url):
        self.base_url = base_url
```

### Objects

```python
client1 = APIClient("service1.com")
client2 = APIClient("service2.com")
```

Both use:

```python
client1.timeout  # 5
client2.timeout  # 5
```

✔ Good use: shared config

---

# 7. Step 7 — Changing class attribute at runtime

```python
APIClient.timeout = 10
```

Now ALL objects see:

```python
client1.timeout  # 10
client2.timeout  # 10
```

👉 This affects all instances

# 8. Step 8 — The biggest bug (VERY IMPORTANT)

## Mutable class attribute bug

```python
class BadService:
    errors = []   # class attribute (shared)
```

```python
s1 = BadService()
s2 = BadService()

s1.errors.append("error1")
```

Now:

```python
s2.errors  # ["error1"]  ❌ unexpected
```

---

## Why this happens

Because:
👉 Both objects use the **same list in class**

---

## Correct way

```python
class GoodService:
    def __init__(self):
        self.errors = []  # instance attribute
```

Now each object is independent ✔

---

# 9. Step 9 — Backend rule (must remember)

👉 Never use mutable class attributes for:

* request data
* user data
* temporary state

### Safe for class attributes:

* constants
* configs
* read-only values

---

# 10. Step 10 — Real production thinking

## Use class attribute for:

* timeouts
* retry counts
* constant configs
* shared metadata

## Use instance attribute for:

* request-specific data
* DB results
* user/session info
* per-object state

❌ Wrong

It is **shared**, not copied.

## Trap 2: modifying mutable class attribute

Very common real-world bug.

# 12. Step 12 — One clean example (interview-ready)

```python
class Cache:
    default_ttl = 300  # class attribute

    def __init__(self):
        self.local_cache = {}  # instance attribute
```

### Explanation:

* `default_ttl` → shared config
* `local_cache` → per-object data

---

# 13. Step 13 — 20-second interview answer

> Instance attributes belong to individual objects and store object-specific data, while class attributes are shared across all instances and typically used for constants or configuration. Python first checks instance attributes, and if not found, falls back to class attributes. A common mistake is using mutable class attributes like lists or dicts, which leads to shared state bugs across instances.

# 14. Step 14 — Follow-up questions

## Q1. What happens if both instance and class have same attribute?

👉 Instance attribute takes priority

---

## Q2. Where should we store configuration?

👉 Class attribute (if shared across instances)

---

## Q3. Why avoid mutable class attributes?

👉 Because they are shared → causes unintended side effects

# 15. Final memory summary (very important)

👉 Instance attribute = **separate per object**
👉 Class attribute = **shared across all objects**

👉 Lookup order:

```
object → class
```

👉 Biggest danger:

```
mutable class attribute → shared bug
```
