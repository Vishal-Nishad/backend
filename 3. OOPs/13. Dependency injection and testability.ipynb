# PART 1 — Dependency Injection (DI)

---

# 1. Step 1 — Core idea (lock this first)

> Dependency Injection means **passing dependencies into a class instead of creating them inside the class**.

---

### Memory line

👉 “**Don’t create dependencies, receive them**”

---

# 2. Step 2 — First understand “dependency”

A dependency is something your class **needs to work**.

Example:

* DB repository
* cache
* API client
* logger

---

# 3. Step 3 — WITHOUT Dependency Injection (BAD)

```python
class UserService:
    def __init__(self):
        self.repo = UserRepository()  # hardcoded
```

---

## Problems

❌ tightly coupled
❌ hard to test
❌ cannot replace repo

---

# 4. Step 4 — WITH Dependency Injection (GOOD)

```python
class UserService:
    def __init__(self, repo):
        self.repo = repo
```

---

### Usage

```python
repo = UserRepository()
service = UserService(repo)
```

---

## What changed?

👉 Service does NOT create repo
👉 Repo is injected

---

# 5. Step 5 — Real backend example (VERY IMPORTANT)

```python
class UserService:
    def __init__(self, repo, cache):
        self.repo = repo
        self.cache = cache

    def get_user(self, user_id):
        user = self.cache.get(user_id)
        if user:
            return user

        user = self.repo.get_by_id(user_id)
        self.cache.set(user_id, user)
        return user
```

---

👉 Dependencies:

* repo
* cache

👉 Injected → flexible system

# 11. Step 11 — Solution WITH DI

```python
class FakeRepo:
    def get_by_id(self, user_id):
        return {"id": user_id, "name": "test"}
```

---

```python
service = UserService(FakeRepo())
```

---

👉 Now:

✔ no DB
✔ fast
✔ predictable

# 12. Step 12 — Real test example

```python
def test_get_user():
    class FakeRepo:
        def get_by_id(self, user_id):
            return {"id": user_id}

    class FakeCache:
        def get(self, key):
            return None
        def set(self, key, value):
            pass

    service = UserService(FakeRepo(), FakeCache())

    result = service.get_user(1)

    assert result["id"] == 1
```

---

👉 This is **real backend testing pattern**

# 18. Step 18 — 30-sec interview answer

> Dependency Injection is a design pattern where dependencies are passed into a class instead of being created inside it. This reduces coupling and makes the system more flexible and testable. In backend systems, it is commonly used for injecting repositories, caches, and external service clients. It improves testability by allowing us to replace real dependencies with mock or fake implementations during testing.
