# PART 1 — `super()` (VERY IMPORTANT)

---

# 1. Step 1 — Core idea (lock this)

> `super()` is used to call methods of the parent (base) class.

---

# 2. Step 2 — Basic example

```python
class Base:
    def greet(self):
        print("Hello from Base")

class Child(Base):
    def greet(self):
        super().greet()
        print("Hello from Child")
```

```python
c = Child()
c.greet()
```

### Output

```
Hello from Base
Hello from Child
```

---

# 3. Step 3 — Why `super()` is needed

Without `super()`:

```python
class Child(Base):
    def greet(self):
        print("Hello from Child")
```

👉 Parent logic is lost ❌

---

# 4. Step 4 — Most important use: constructor chaining

This is **VERY commonly asked**

```python
class Base:
    def __init__(self, name):
        self.name = name

class Child(Base):
    def __init__(self, name, age):
        super().__init__(name)   # important
        self.age = age
```

---

## What happens

```python
c = Child("Vishal", 25)
```

Flow:

1. Child `__init__` runs
2. Calls `super().__init__(name)`
3. Base initializes `name`

---

## Without super (bug)

```python
class Child(Base):
    def __init__(self, name, age):
        self.age = age
```

👉 `name` is never set → bug ❌

---

# 5. Step 5 — Backend example (important)

```python
class BaseRepository:
    def __init__(self, session):
        self.session = session

class UserRepository(BaseRepository):
    def __init__(self, session):
        super().__init__(session)
```

👉 Ensures DB session is initialized correctly

---

# 6. Step 6 — Mental model (easy memory)

👉 `super()` = “go to parent and reuse its logic”

---

# 7. Step 7 — Interview traps

## Trap 1: forgetting super in constructor

👉 very common bug

---

## Trap 2: calling parent manually

```python
Base.__init__(self, name)  ❌
```

👉 Works but not recommended
👉 Use `super()`

---

## Trap 3: thinking super only for init

❌ Wrong

# 8. Step 8 — 20-sec interview answer

> `super()` is used to call parent class methods, especially constructors. It helps reuse parent logic and ensures proper initialization in inheritance. It’s commonly used in backend code when extending base classes like repositories or services.

# PART 2 — Magic / Dunder Methods

---

# 9. Step 9 — Core idea (lock this)

> Magic (dunder) methods are special methods with double underscores that define how objects behave.

---

## Examples:

* `__init__`
* `__str__`
* `__repr__`
* `__len__`

---

# 10. Step 10 — Most important ones (only what matters)

---

## 1. `__init__` (already covered)

👉 initializes object

---

## 2. `__str__` (VERY useful)

👉 defines how object prints

```python
class User:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"User({self.name})"
```

```python
print(User("Vishal"))
```

Output:

```
User(Vishal)
```

---

## Backend use

* logging
* debugging
* readable output

---

## 3. `__repr__` (interview favorite)

👉 developer-friendly representation

```python
def __repr__(self):
    return f"User(name={self.name})"
```

### Difference

| Method     | Purpose         |
| ---------- | --------------- |
| `__str__`  | user-friendly   |
| `__repr__` | developer/debug |

---

## Interview line

> `__repr__` should ideally be unambiguous and useful for debugging.

---

## 4. `__len__`

```python
class Cart:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)
```

```python
len(cart)
```

---

## 5. `__eq__` (important for comparisons)

```python
class User:
    def __init__(self, id):
        self.id = id

    def __eq__(self, other):
        return self.id == other.id
```

---

## Backend use

* comparing DB entities
* caching keys
* testing

---

# 11. Step 11 — What you should NOT focus on

Skip deep study of:

* `__new__`
* `__getattr__`, `__setattr__`
* operator overloading in depth

👉 low ROI for backend interviews

# 12. Step 12 — Real backend example

```python
class APIResponse:
    def __init__(self, status, data):
        self.status = status
        self.data = data

    def __repr__(self):
        return f"APIResponse(status={self.status}, data={self.data})"
```

👉 useful for logs/debugging

---

# 13. Step 13 — Common mistakes

## Mistake 1: not implementing `__repr__`

👉 debugging becomes hard

---

## Mistake 2: wrong equality logic

```python
return self.id == other.id
```

👉 must handle `other` type safely

---

## Mistake 3: overusing dunder methods

👉 unnecessary complexity

---

# 14. Step 14 — Memory shortcut

👉 `super()` → parent logic
👉 `__init__` → initialize
👉 `__str__` → print
👉 `__repr__` → debug
👉 `__len__` → length
👉 `__eq__` → compare

---

# 15. Step 15 — Final 30-sec interview answer

> `super()` is used to call parent class methods and is commonly used in constructors to ensure proper initialization in inheritance. Magic or dunder methods are special methods like `__init__`, `__str__`, `__repr__`, and `__eq__` that define how objects behave in Python. In backend systems, they are mainly useful for initialization, debugging, logging, and comparisons.

---

# 16. One line to impress interviewer

> I mainly use `super()` for constructor chaining and rely on key dunder methods like `__repr__` and `__eq__` to improve debugging and correctness in backend systems, while avoiding unnecessary magic methods.
