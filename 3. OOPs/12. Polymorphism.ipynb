# 4. Step 4 — Production example (VERY IMPORTANT)

## Payment system

### Abstraction

```python
class PaymentGateway:
    def charge(self, amount):
        pass
```

---

### Implementations

```python
class StripeGateway(PaymentGateway):
    def charge(self, amount):
        return "Stripe payment"

class RazorpayGateway(PaymentGateway):
    def charge(self, amount):
        return "Razorpay payment"
```

---

### Service

```python
class PaymentService:
    def __init__(self, gateway):
        self.gateway = gateway

    def pay(self, amount):
        return self.gateway.charge(amount)
```

---

### Usage

```python
service = PaymentService(StripeGateway())
service.pay(100)
```

---

## Key insight

👉 `PaymentService` doesn’t care:

* Stripe or Razorpay

👉 It only cares:

```plaintext
object has charge() method
```

# 6. Step 6 — Types of polymorphism (only what matters)

## 1. Method overriding (MOST IMPORTANT)

```python
class Base:
    def process(self):
        pass

class A(Base):
    def process(self):
        return "A"

class B(Base):
    def process(self):
        return "B"
```

---

## 2. Duck typing (VERY PYTHONIC)

```python
class FileLogger:
    def log(self, msg):
        print(msg)

class DBLogger:
    def log(self, msg):
        print(msg)
```

```python
def log_message(logger):
    logger.log("hello")
```

👉 No inheritance needed ✔

# 1. Step 1 — Core idea (lock this first)

> Duck typing means:
> 👉 **If an object behaves like what we expect, we don’t care about its type**

---

### Famous line

👉 “**If it looks like a duck and quacks like a duck, it’s a duck**”

---

# 2. Step 2 — What this means in Python

In Python:

👉 We don’t check type
👉 We check behavior (methods/attributes)

---

# 3. Step 3 — Simple example

```python
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"
```

```python
def make_sound(animal):
    return animal.speak()
```

```python
make_sound(Dog())
make_sound(Cat())
```

---

### Important point

👉 No inheritance
👉 No base class
👉 Still works ✔

---

# 4. Step 4 — Why this works

Because both objects have:


👉 That’s all Python cares about

# 7. Step 7 — Duck typing vs inheritance

| Feature             | Duck Typing | Inheritance |
| ------------------- | ----------- | ----------- |
| Requires base class | ❌ No        | ✔ Yes       |
| Flexibility         | High        | Medium      |
| Strictness          | Low         | High        |

# 8. Step 8 — Duck typing vs abstraction (important)

| Concept           | Meaning           |
| ----------------- | ----------------- |
| Abstraction (ABC) | explicit contract |
| Duck typing       | implicit contract |

---

### Example

#### With ABC (strict)

```python
class PaymentGateway(ABC):
    @abstractmethod
    def charge(self): pass
```

---

#### With duck typing (loose)

```python
class FakeGateway:
    def charge(self):
        return "fake"
```

👉 Works without inheritance ✔

# 11. Step 11 — Common mistake (IMPORTANT)

## Mistake: assuming method exists

```python
def process(obj):
    obj.run()
```

If `run()` not present → runtime error ❌

---

## Safer way

```python
if hasattr(obj, "run"):
    obj.run()
```

```python
try:
    obj.run()

    pass
```
