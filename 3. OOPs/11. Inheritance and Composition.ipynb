# 1. Step 1 — Core idea (lock this first)

> Inheritance means a child class **reuses and extends** a parent class.

---

### Memory line

👉 “**is-a relationship + reuse behavior**”

Example:

* `UserRepository` **is a** `BaseRepository`

---

# 2. Step 2 — Basic example (simple but important)

```python
class BaseRepository:
    def save(self, obj):
        print("Saving to DB")

class UserRepository(BaseRepository):
    def get_user(self, user_id):
        print("Fetching user")
```

```python
repo = UserRepository()
repo.save("data")       # from parent
repo.get_user(1)        # own method
```

---

## What happened?

* Child **inherits** parent methods
* No need to rewrite `save()`

---

# 3. Step 3 — Why inheritance is used

## Main reasons:

### 1. Code reuse

Avoid duplicating logic

### 2. Shared behavior

Common functionality in base class

### 3. Structure

Organize related classes

# 4. Step 4 — Most important concept: method overriding

Child modifies parent behavior.

```python
class BaseCache:
    def get(self, key):
        return "base"

class RedisCache(BaseCache):
    def get(self, key):
        return f"redis:{key}"
```

👉 Same method name, different behavior

---

## Interview line

> Overriding allows child classes to provide specific implementations while keeping a common interface.

# 5. Step 5 — Using `super()` (very important)

```python
class BaseService:
    def process(self):
        print("Base logic")

class UserService(BaseService):
    def process(self):
        super().process()
        print("User logic")
```

👉 Reuse + extend

# 6. Step 6 — Constructor inheritance (VERY COMMON QUESTION)

```python
class Base:
    def __init__(self, db):
        self.db = db

class Child(Base):
    def __init__(self, db, cache):
        super().__init__(db)
        self.cache = cache
```

👉 Always call `super().__init__`

---

## Interview trap

Forgetting `super()` → parent not initialized ❌

> Composition means a class **uses other classes as dependencies** instead of inheriting from them.

# 2. Step 2 — Basic example

```python id="6wew8b"
class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self):
        self.engine = Engine()  # composition

    def start(self):
        return self.engine.start()
```

👉 Car **has an** Engine
👉 Not “Car is Engine”

# 4. Step 4 — Real production example (VERY IMPORTANT)

## User service with DB + Cache

```python id="54w2yu"
class UserRepository:
    def get_by_id(self, user_id):
        return {"id": user_id, "name": "Vishal"}

class Cache:
    def get(self, key):
        return None

    def set(self, key, value):
        pass


class UserService:
    def __init__(self, repo, cache):
        self.repo = repo      # composition
        self.cache = cache    # composition

    def get_user(self, user_id):
        user = self.cache.get(user_id)
        if user:
            return user

        user = self.repo.get_by_id(user_id)
        self.cache.set(user_id, user)
        return user
```

---

## Why this is powerful

* Replace cache easily (Redis → in-memory)
* Replace DB easily
* Easy to test (mock repo/cache)
