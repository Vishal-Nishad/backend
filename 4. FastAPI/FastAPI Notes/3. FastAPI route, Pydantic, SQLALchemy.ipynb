# PART 1 ŌĆö What is SQLAlchemy?

## ­¤ö╣ SQLAlchemy = Python DB toolkit / ORM library

It helps Python talk to databases like:

* PostgreSQL
* MySQL
* SQLite

Without writing raw SQL everywhere.

---

## Without SQLAlchemy:

You manually write SQL:

```sql id="2ebqpk"
INSERT INTO product (name, price) VALUES ('phone', 99);
```

---

## With SQLAlchemy:

You write Python:

```python id="bb2bze"
product = Product(name="phone", price=99)
db.add(product)
db.commit()
```

­¤æē SQLAlchemy converts this into SQL behind the scenes.

---

# Production meaning

In real projects, SQLAlchemy is used because:

* Cleaner than raw SQL everywhere
* Easier to manage models
* Integrates well with FastAPI
* Supports relationships, transactions, migrations, etc.

---

# PART 2 ŌĆö What is ORM?

## ORM = Object Relational Mapping

This is a **very important interview topic**.

---

## Break the term:

### Object

Python object:

```python id="r1s6qk"

```

### Relational

Database table:

```sql id="0w2go3"

```

### Mapping

ORM maps:

* Python class Ōåö DB table
* Python object Ōåö DB row

---

# ­¤öź Example

## Python object:

```python id="k2m0el"

```

## DB row:

```sql id="g0id1g"

1  | phone | 99
```

­¤æē ORM says:

> ŌĆ£These are the same thing in two worlds.ŌĆØ

---

# ­¤¦Ā Best mental model

­¤æē ORM = **translator between Python objects and SQL tables**

# PART 3 ŌĆö What is `declarative_base()`?

Now your code:

```python id="z2c0kh"

```

This line confuses almost everyone initially.

---

## What it means simply:

`Base` is the **parent class** for all your DB models.

---

### Like this:

```python id="o9mrh5"

    ...
```

This tells SQLAlchemy:

> ŌĆ£This class is not a normal Python class only ŌĆö it is a DB model.ŌĆØ

---

## Why needed?

Because SQLAlchemy needs to track:

* table names
* columns
* metadata
* relationships

---

## Best mental model

­¤æē `Base` = **registry / foundation for all DB tables**

---

# Example

```python id="78muec"
class Product(Base):
    __tablename__ = "product"
    id = Column(Integer, primary_key=True)
```

SQLAlchemy stores internally:

* table name = `product`
* columns = `id`, `name`, etc.

So later it can create table in DB.
